```python
import pandas as pd
import os

# Paths
processed_path = r"C:\Users\ruham\Downloads\financial inclusion forecasting system\Financial-inclusion-forecasting-system-week10\data\processed"
file_name = "ethiopia_fi_unified_data_enriched.xlsx"

# Load dataset
df = pd.read_excel(os.path.join(processed_path, file_name))

# Quick overview
print(df.shape)
print(df.columns)
df.head()
```

```python
# Info & types
df.info()

# Missing values
print(df.isna().sum())

# Quick statistics
df.describe(include='all')
```

```python
# Year and month from observation_date
df['year'] = df['observation_date'].dt.year
df['month'] = df['observation_date'].dt.month
df['quarter'] = df['observation_date'].dt.quarter
```

```python
# normalize value_numeric by pillar
df['value_norm'] = df.groupby('pillar')['value_numeric'].transform(lambda x: (x - x.mean()) / x.std())
```

```python
# Convert record_type and source_type to category
df['record_type'] = df['record_type'].astype('category')
df['source_type'] = df['source_type'].astype('category')
```

```python
# Ensure date is datetime
df['observation_date'] = pd.to_datetime(df['observation_date'])

# Year already exists, but this is safe
df['year'] = df['observation_date'].dt.year

# Split by pillar
access_df = df[df['pillar'] == 'ACCESS']
usage_df = df[df['pillar'] == 'USAGE']

# Compute yearly trends
access_trend = access_df.groupby('year')['value_numeric'].mean()
usage_trend = usage_df.groupby('year')['value_numeric'].mean()

# Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(access_trend.index, access_trend.values, label='ACCESS', marker='o')
plt.plot(usage_trend.index, usage_trend.values, label='USAGE', marker='o')
plt.title('ACCESS vs USAGE Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Average Indicator Value')
plt.legend()
plt.grid(True)
plt.show()
```

### Insights
- ACCESS indicators show a steady upward trend, suggesting improved financial availability.
- USAGE indicators grow more slowly, indicating adoption lags behind access.


```python
access_yoy = access_trend.pct_change() * 100
usage_yoy = usage_trend.pct_change() * 100

plt.figure()
plt.plot(access_yoy.index, access_yoy.values)
plt.title("ACCESS Growth Rate (%)")
plt.show()
```

- Growth slows after certain years, indicating possible market saturation.
- Volatility increases around policy or economic shifts.


```python
events = df[df['record_type'] == 'event']

for _, row in events.iterrows():
    plt.axvline(row['year'], linestyle='--', alpha=0.3)
```

- Growth fluctuations align with major economic or policy events.
- Some indicators recover faster than others post-event.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Observations per year
sns.countplot(x='year', data=df)
plt.title('Observations per Year')
plt.show()

# Observations per pillar
sns.countplot(x='pillar', data=df)
plt.title('Observations per Pillar')
plt.show()
```

```python
# Numeric correlations
corr = df[['value_numeric', 'value_norm']].corr()
print(corr)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
